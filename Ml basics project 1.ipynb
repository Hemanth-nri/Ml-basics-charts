```python
from matplotlib import pyplot as plt
```

```python

import random
x = []
y = []
#loop comprehension
[x.append(random.randint(5,25)) for _ in range(10)]
[y.append(random.randint(5,25)) for _ in range(10)]
#print the x and y values
print(x)
print(y)
```

```python
#pyplot - method to plot the x and y values
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.bar(x,y)
```
